# def GraficoLinha(style)

```python
def GraficoLinha(style = 'ggplot'):
  #estilo de cor
  plt.style.use(style) 

  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico
  plt.show()
```

```python
GraficoLinha('dark_background')
```

```python
df = pd.read_csv('Idade_Carro.csv')
df
```

```python
def graficoDispersao(style = 'ggplot'):
  # Criar o gráfico de dispersão
  plt.style.use(style)
  plt.scatter(df['Idade'], df['Preço'])

  plt.title('Preço do Carro x Idade')
  plt.xlabel('Idade do Carro (Anos)')
  plt.ylabel('Preço (R$)')

  # Mostrar o gráfico
  plt.show()
```

```python
graficoDispersao('dark_background')
```

# Explique como esse gráfico:

# Por que é interessante fazer um gráfico de dispersão nesse caso?
R: relacionando a idade e o valor do caro é possível determinar a curva de queda do preço conforme o tempo

# Qual é a diferença entre o gráfico de linha na interpretação?
R: a conexão entre os pontos no gráfico de linha determina que o valor flutua no período intermediário entre os anos



# Historiograma

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição e compare a diferença com a distribuição normal

R: a diferença da distribuição para normal, é que a distribuição normal apresenta valores minimos e máximos com menos frequência, enquanto a distribuição aleatória fica mais homogênea

# Gráfico de Donut

```python
# Dados
generos = ['Feminino', 'Masculino']
quantidade = [650, 550]

#Porcentagem
for i in range(0, len(generos)):
  generos[i] = generos[i] + ' - ' + str(round(quantidade[i]*100/sum(quantidade), 1)) + '%'

# Criar o gráfico de pizza
plt.pie(quantidade, labels=generos, startangle=90, colors=['#FFC0CB', '#0000FF'], wedgeprops={'width': 0.4},
        textprops={'fontsize': 20})

plt.title('Distribuição de Gênero', fontdict={'fontsize': 20})


# Mostrar o gráfico
plt.show()
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)
plt.style.use('ggplot')
# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: uma distribuição assimétrica negativa, visto que a mediana é mais próxima do terceiro quartil

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```
